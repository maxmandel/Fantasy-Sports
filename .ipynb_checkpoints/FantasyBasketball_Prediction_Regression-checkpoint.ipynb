```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import kstest

from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.ar_model import AutoReg

from nba_api.stats.endpoints import playergamelog, playerprofilev2
from nba_api.stats.static import players
```

```python
def get_player_gamelogs_all_seasons(player_name):
    
    player_info = players.find_players_by_full_name(player_name)
    if not player_info:
        raise ValueError(f"Player {player_name} not found.")
    
    player_id = player_info[0]['id']
    
    # Fetch player profile to determine rookie season
    profile = playerprofilev2.PlayerProfileV2(player_id=player_id)
    seasons_stats = profile.season_totals_regular_season.get_data_frame()
    rookie_season = int(seasons_stats['SEASON_ID'].min()[:4])
    current_season = int(seasons_stats['SEASON_ID'].max()[:4]) + 1

    seasons = [f"{year}-{str(year + 1)[-2:]}" for year in range(rookie_season, current_season)]
    
    # Fetch game logs for each season
    all_season_gamelogs = []
    for season in seasons:
        try:
            gamelog = playergamelog.PlayerGameLog(player_id=player_id, season=season)
            season_df = gamelog.get_data_frames()[0]
            season_df['SEASON'] = season
            all_season_gamelogs.append(season_df)
        except Exception as e:
            print(f"Error retrieving data for season {season}: {e}")

    # Combine all DataFrames
    if all_season_gamelogs:
        return pd.concat(all_season_gamelogs, ignore_index=True)
    else:
        return pd.DataFrame()

def calculate_fantasy_points(row):
    
    points = row['PTS'] * 0.5
    rebounds = row['REB']
    assists = row['AST']
    steals = row['STL'] * 2
    blocks = row['BLK'] * 2
    turnovers = row['TOV'] * -1
    threes_made = row['FG3M'] * 0.5

    fantasy_points = points + rebounds + assists + steals + blocks + turnovers + threes_made

    stat_categories = sum(stat >= 10 for stat in [row['PTS'], row['REB'], row['AST'], row['STL'], row['BLK']])
    if stat_categories >= 2:
        fantasy_points += 1  # Double-double bonus
    if stat_categories >= 3:
        fantasy_points += 2  # Triple-double bonus (includes double-double bonus)

    return fantasy_points

def get_all_fpts(player_name):
    
    df = get_player_gamelogs_all_seasons(player_name)
    all_fpts = []
    game_dates = []
    
    for _, row in df.iterrows():
        fpts = calculate_fantasy_points(row)
        all_fpts.append(fpts)
        game_dates.append(row['GAME_DATE'])
        
    return len(all_fpts), all_fpts, game_dates
```

```python
def plot_history(player_name):
    n, fpts_arr, dates = get_all_fpts(player_name)
    dates_dt = pd.to_datetime(dates, format='%b %d, %Y')

    sorted_indices = np.argsort(dates_dt)  
    dates_dt = dates_dt[sorted_indices]  
    fpts_arr = np.array(fpts_arr)[sorted_indices]  

    seasons = np.array([(d.year if d.month >= 10 else d.year - 1) for d in dates_dt])

    unique_seasons = np.unique(seasons)
    color_map = plt.colormaps.get_cmap('viridis')

    plt.figure(figsize=(10, 6))

    for i, season in enumerate(unique_seasons):
        season_mask = (seasons == season)
        season_dates = dates_dt[season_mask]
        season_fpts = fpts_arr[season_mask]
        
        plt.scatter(dates_dt[season_mask], fpts_arr[season_mask], 
                    color=color_map(i / len(unique_seasons)),
                    label=f"{season}-{season+1}", s=20)
        
        if len(season_dates) > 3:  # Fit only if enough points exist for a cubic fit
            season_x = (season_dates - season_dates.min()).days  # Convert to numerical days
            trend_coeffs = np.polyfit(season_x, season_fpts, 3)  # **Degree 3 fit**
            trend_poly = np.poly1d(trend_coeffs)
            smooth_x = np.linspace(season_x.min(), season_x.max(), 100)  # Smooth curve
            plt.plot(season_dates.min() + pd.to_timedelta(smooth_x, unit='D'), 
                     trend_poly(smooth_x), color='red', linestyle='-', linewidth=2)  # **Cubic Trendline**

        season_avg = np.mean(season_fpts)
        plt.hlines(y=season_avg, xmin=season_dates.min(), xmax=season_dates.max(), colors='blue', linestyle='-', linewidth=2)

    if len(dates_dt) > 3:
        overall_x = (dates_dt - dates_dt.min()).days  # Convert dates to numerical days
        overall_trend_coeffs = np.polyfit(overall_x, fpts_arr, 3)  # **Degree 3 fit**
        overall_trend_poly = np.poly1d(overall_trend_coeffs)
        smooth_x = np.linspace(overall_x.min(), overall_x.max(), 300)  # Smooth curve
        plt.plot(dates_dt.min() + pd.to_timedelta(smooth_x, unit='D'), 
                 overall_trend_poly(smooth_x), color='green', linestyle='-', linewidth=2, label="Overall Trend")  # **Cubic Trendline**
    
    plt.plot(dates_dt, fpts_arr, linestyle='-', color='gray', alpha=0.5)

    plt.title(f"{player_name}: Fantasy Points Over Time")
    plt.xlabel("Game Date")
    plt.ylabel("Fantasy Points")
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
    plt.tight_layout()
    plt.show()
```

```python
plot_history("Josh Hart")
```

```python
def plot_autocorrelation_pacf(player_name, max_lag=20):

    n, fpts_arr, dates = get_all_fpts(player_name)
    dates_dt = pd.to_datetime(dates, format='%b %d, %Y')

    sorted_indices = np.argsort(dates_dt)  
    fpts_arr = np.array(fpts_arr)[sorted_indices]  

    lags = np.arange(max_lag + 1)
    autocorr_values = acf(fpts_arr, nlags=max_lag, fft=True)
    pacf_values = pacf(fpts_arr, nlags=max_lag)

    # Create subplots
    fig, axes = plt.subplots(2, 1, figsize=(10, 10))

    # ACF
    axes[0].bar(lags, autocorr_values, color='blue', alpha=0.7, label="ACF")
    axes[0].axhline(y=0, color='black', linewidth=1)
    axes[0].axhline(y=1.96/np.sqrt(len(fpts_arr)), color='red', linestyle='dashed', linewidth=1)
    axes[0].axhline(y=-1.96/np.sqrt(len(fpts_arr)), color='red', linestyle='dashed', linewidth=1)
    axes[0].set_xlabel("Lag (Games)")
    axes[0].set_ylabel("Autocorrelation")
    axes[0].set_title(f"{player_name}: Autocorrelation Function (ACF)")
    axes[0].legend()
    axes[0].grid(True)

    # PACF
    axes[1].bar(lags, pacf_values, color='green', alpha=0.7, label="PACF")
    axes[1].axhline(y=0, color='black', linewidth=1)
    axes[1].axhline(y=1.96/np.sqrt(len(fpts_arr)), color='red', linestyle='dashed', linewidth=1)
    axes[1].axhline(y=-1.96/np.sqrt(len(fpts_arr)), color='red', linestyle='dashed', linewidth=1)
    axes[1].set_xlabel("Lag (Games)")
    axes[1].set_ylabel("Partial Autocorrelation")
    axes[1].set_title(f"{player_name}: Partial Autocorrelation Function (PACF)")
    axes[1].legend()
    axes[1].grid(True)

    # Show plots
    plt.tight_layout()
    plt.show()
```

```python
plot_autocorrelation_pacf("Ivica Zubac")
```

```python
def fit_ar5_predict(player_name):

    n, fpts_arr, dates = get_all_fpts(player_name)
    dates_dt = pd.to_datetime(dates, format='%b %d, %Y')
    
    sorted_indices = np.argsort(dates_dt)  
    dates_dt = dates_dt[sorted_indices]  
    fpts_arr = np.array(fpts_arr)[sorted_indices]  

    ar_model = AutoReg(fpts_arr, lags=5).fit()

    next_fpt = ar_model.predict(start=len(fpts_arr), end=len(fpts_arr))[0]
    
    residuals = fpts_arr[5:] - ar_model.fittedvalues
    residual_std = np.std(residuals)
    conf_interval = 1.96 * residual_std  # 95% CI

    lower_bound = next_fpt - conf_interval
    upper_bound = next_fpt + conf_interval

    print(f"Predicted Fantasy Points for Next Game: {next_fpt:.2f}")
    print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    plt.figure(figsize=(10, 6))
    plt.plot(dates_dt, fpts_arr, marker='o', linestyle='-', label="Actual Fantasy Points")
    plt.plot(dates_dt[5:], ar_model.fittedvalues, linestyle='dashed', color='red', label="AR(5) Fitted Values")

    next_game_date = dates_dt[-1] + pd.Timedelta(days=3)
    plt.scatter(next_game_date, next_fpt, color='green', label="1-Game Ahead Prediction", zorder=3, s=100)
    plt.vlines(next_game_date, lower_bound, upper_bound, color='green', linestyle='dashed', linewidth=2, label="95% CI")

    plt.title(f"{player_name}: AR(5) Model and 1-Game Prediction with 95% CI")
    plt.xlabel("Game Date")
    plt.ylabel("Fantasy Points")
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.legend()
    plt.show()

    return next_fpt, (lower_bound, upper_bound)


predicted_fpts, confidence_interval = fit_ar5_predict("Ivica Zubac")
```
